```python
import numpy as np
import matplotlib.pyplot as plt
# Definir parmetros
fs = 1000          # Frecuencia de muestreo (Hz)
t = np.linspace(0, 1, fs)   # Vector de tiempo de 0 a 1 segundo
f = 5              # Frecuencia de la seal (Hz)

# Crear seal senoidal
senal = np.sin(2 * np.pi * f * t)

# Graficar la seal
plt.plot(t, senal)
plt.title("Seal senoidal en el dominio del tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

Creamos una seal senoidal de 5 Hz muestreada a 1000 Hz.  
Graficamos la seal en el dominio del tiempo para observar su forma.

```python
# Calcular la Transformada de Fourier
fft_senal = np.fft.fft(senal)

# Obtener el eje de frecuencias
frecuencias = np.fft.fftfreq(len(senal), 1/fs)

# Calcular la magnitud
magnitud = np.abs(fft_senal)

# Graficar solo la parte positiva de las frecuencias
plt.plot(frecuencias[:fs//2], magnitud[:fs//2])
plt.title("Espectro de magnitud")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.grid()
plt.show()
```

Aplicamos la FFT usando `np.fft.fft()` para calcular el espectro de frecuencia.  
Graficamos la magnitud y la fase.  
La magnitud muestra un pico en 5 Hz, y la fase indica el desfase de la seal.

```python
# Calcular la fase
fase = np.angle(fft_senal)

# Graficar fase
plt.plot(frecuencias[:fs//2], fase[:fs//2])
plt.title("Espectro de fase")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Fase (radianes)")
plt.grid()
plt.show()
```

Creamos una segunda seal senoidal de 20 Hz y combinamos ambas seales con coeficientes a=2 y b=3.  
Comparando la FFT de la combinacin con la combinacin de las FFTs individuales, observamos que coinciden.  
Esto verifica la propiedad de linealidad

```python
# Crear segunda seal
f2 = 20  # frecuencia 20 Hz
senal2 = np.sin(2 * np.pi * f2 * t)

# Crear combinacin lineal
a = 2
b = 3
senal_combinada = a * senal + b * senal2

# FFT de cada seal
fft1 = np.fft.fft(senal)
fft2 = np.fft.fft(senal2)
fft_combinada = np.fft.fft(senal_combinada)

# Verificar propiedad de linealidad
fft_linealidad = a * fft1 + b * fft2

# Graficar magnitud de ambas
plt.plot(frecuencias[:fs//2], np.abs(fft_combinada[:fs//2]), label="FFT directa")
plt.plot(frecuencias[:fs//2], np.abs(fft_linealidad[:fs//2]), '--', label="Propiedad linealidad")
plt.title("Verificacin de Linealidad")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.legend()
plt.grid()
plt.show()
```

Retrasamos la seal original 0.1 s usando `np.roll`.  
La magnitud del espectro se mantiene igual, pero la fase cambia, demostrando la propiedad de desplazamiento en el tiempo.

```python
# Desplazamiento en el tiempo (retraso)
delay = 0.1  # segundos
samples_delay = int(delay * fs)  # cantidad de muestras a retrasar

# Seal desplazada (usamos np.roll para desplazar muestras)
senal_desplazada = np.roll(senal, samples_delay)

# FFT de la seal desplazada
fft_desplazada = np.fft.fft(senal_desplazada)

# Magnitud y fase
magnitud_desplazada = np.abs(fft_desplazada)
fase_desplazada = np.angle(fft_desplazada)

# Graficar magnitud original vs desplazada
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(frecuencias[:fs//2], magnitud[:fs//2], label="Magnitud original")
plt.plot(frecuencias[:fs//2], magnitud_desplazada[:fs//2], '--', label="Magnitud desplazada")
plt.title("Magnitud: original vs desplazada")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.legend()
plt.grid()

# Graficar fase original vs desplazada
plt.subplot(1,2,2)
plt.plot(frecuencias[:fs//2], fase[:fs//2], label="Fase original")
plt.plot(frecuencias[:fs//2], fase_desplazada[:fs//2], '--', label="Fase desplazada")
plt.title("Fase: original vs desplazada")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Fase (radianes)")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

Comprimimos la seal en el tiempo con un factor de 2.  
La frecuencia se duplica y la magnitud del espectro muestra el pico en 10 Hz, cumpliendo la propiedad de escalamiento.

```python
factor = 2  # >1 comprime la seal, <1 la estira
t_scaled = t / factor
senal_scaled = np.sin(2 * np.pi * f * t_scaled)

# FFT de la seal escalada
fft_scaled = np.fft.fft(senal_scaled)
magnitud_scaled = np.abs(fft_scaled)

# Graficar magnitud original vs escalada
plt.plot(frecuencias[:fs//2], magnitud[:fs//2], label="Original")
plt.plot(frecuencias[:fs//2], magnitud_scaled[:fs//2], '--', label="Seal escalada")
plt.title("Propiedad de escalamiento en frecuencia")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.legend()
plt.grid()
plt.show()
```

En esta seccin graficamos lado a lado la magnitud y fase de seales combinadas o desplazadas.  
Esto nos permite visualizar claramente cmo cambian la fase y cmo se mantiene la magnitud, comprobando las propiedades de la Transformada de Fourier.

```python
# Pulso rectangular
senal_rect = np.where((t >= 0.3) & (t <= 0.7), 1, 0)

# Graficar la seal
plt.plot(t, senal_rect)
plt.title("Pulso rectangular en el dominio del tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

El pulso rectangular vale 1 en un intervalo de tiempo definido (0.3 a 0.7 s) y 0 fuera de l.
Su espectro de frecuencia contiene muchas componentes, reflejando los cambios bruscos de la seal en el tiempo.a

```python
# Funcin escaln
t0 = 0.5  # tiempo de inicio del escaln
senal_escalon = np.where(t >= t0, 1, 0)

# Graficar la seal
plt.plot(t, senal_escalon)
plt.title("Funcin escaln en el dominio del tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

```python
# FFT de la funcin escaln
fft_escalon = np.fft.fft(senal_escalon)
magnitud_escalon = np.abs(fft_escalon)
fase_escalon = np.angle(fft_escalon)

# Graficar magnitud y fase
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(frecuencias[:fs//2], magnitud_escalon[:fs//2])
plt.title("Magnitud - Funcin escaln")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.grid()

plt.subplot(1,2,2)
plt.plot(frecuencias[:fs//2], fase_escalon[:fs//2])
plt.title("Fase - Funcin escaln")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Fase (radianes)")
plt.grid()

plt.tight_layout()
plt.show()
```

La funcin escaln cambia de 0 a 1 en un instante especfico (t0 = 0.5 s).
Su espectro de frecuencia tiene componentes de baja frecuencia dominantes y fases con saltos,
lo que muestra cmo los cambios abruptos en el tiempo generan muchas frecuencias.
